```python
import numpy as np
from sklearn.naive_bayes import MultinomialNB, BernoulliNB
from sklearn.metrics import accuracy_score
from scipy.sparse import coo_matrix
```

```python
#data
path = '/content/drive/MyDrive/VCCorp/Naive_Bayes_Classifier/ex6DataPrepared/'
train_data_fn = 'train-features.txt'
test_data_fn ='test-features.txt'
train_label_fn = 'train-labels.txt'
test_label_fn = 'test-labels.txt'
```

```python
nwords = 2500

def read_data(data_fn, label_fn):
  #read label_fn
  with open(path+label_fn,'r') as f:
    content = f.readlines()
  # strip() loai bo dau '\n'
  label = [int(x.strip()) for x in content]
  
  # read data_fn
  with open(path + data_fn,'r') as f:
    content = f.readlines()
  content = [x.strip() for x in content]

  dat = np.zeros((len(content), 3), dtype=int)
  for i, line in enumerate(content):
    a = line.split()
    dat[i,:] = np.array([int(a[0]), int(a[1]), int(a[2])])
  # chuyen ve ma tran thua
  data = coo_matrix((dat[:,2], (dat[:,0]-1, dat[:,1]-1)), shape=(len(label), nwords))
  return data, label
```

Mô hình Multinomial

```python
(train_data, train_label) = read_data(train_data_fn, train_label_fn)
(test_data, test_label) = read_data(test_data_fn, test_label_fn)

model = MultinomialNB()
model.fit(train_data, train_label)

y_pred = model.predict(test_data)
print('Training size = ',train_data.shape[0],' Accuracy = %.2f'%(accuracy_score(y_pred, test_label)*100))
```
